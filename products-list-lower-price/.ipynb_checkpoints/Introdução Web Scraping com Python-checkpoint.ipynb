```python
#Imports
from requests import get
from requests.exceptions import RequestException
from contextlib import closing
from bs4 import BeautifulSoup
import pandas
```

```python
#Fazendo as requisições para o download das páginas
#Criando função para receber uma URL e verificar se tem algum 
#Conteúdo do tipo HTML/texto 
def get_url(url):
    try:
        with closing(get(url,stream=True)) as resp:
            if response_true(resp):
                return resp.content
            else:
                return None
    except RequestException as e:
        log_error("Erro de solicitação ".format(url,str(e)))
        return None
    
#Função para imprimir os erros 
def log_error(e):
    print(e)
    
#Função que retorna true caso seja HTML
def response_true(resp):
    content_type = resp.headers['Content-Type'].lower()
    return (resp.status_code==200
           and content_type is not None
           and content_type.find('html')>-1)
page = get_url("https://www.submarino.com.br/categoria/informatica/notebook")
```

```python
#Tratando o HTML bruto
html = BeautifulSoup(page, 'html.parser')

#Pegando todos os H1 da página que tem a classe 'card-product-name' com os detalhes
details = html.findAll("div", {"class": "card-product-details"})
products = []
prices = []
for product in details:
    products.append(product.find("h1", {"class":"card-product-name"}).text)
    prices.append(product.find_all("span", {"class":"value"})[1].text)

#Ordenando do menor para o maior
df = pandas.DataFrame({
    "Product": products,
    "Price": sorted(prices)
})
```
